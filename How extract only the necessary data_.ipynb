```python
all_data = []
scanning = False
for i in open(path).readlines():
    if 'Non-deleted saved meals' in i:
        scanning = True
        continue
    
    if i.startswith('------------------'):
        scanning = False
        continue
    
    if scanning:
        # now we need to deal with columns
        if 'Meal#' in i:
            col_data = i
            continue
        
        # now add necessary data to dataframe
        all_data.append([value.strip() for value in i.split(';')])
```

```python
# since there is some space between names, and ';' in col_data, we need to organize it
final_columns = [x.strip() for x in col_data.split(';')]
final_columns
```

```python
# finish with dataframe
meal_df = pd.DataFrame(all_data, columns=final_columns[0:17])
meal_df
```
